```python
def categorize(value):
    categories = [
        ("A-", 0.0, 49.0),
        ("A1", 48.0, 58.0),
        ("A2", 58.0, 68.0),
        ("B1", 68.0, 82.0),
        ("B+", 79.0, 100.0),
    ]
    for category, min_val, max_val in categories:
        if min_val <= value <= max_val:
            return category
    return "Unknown"  # For values outside defined ranges
```

# Cleaning Zone

```python

personashogar_dict={"Una":"1 a 2",
                    "1 a 2":"1 a 2",
                    "Dos":"1 a 2",
                    "Tres":"3 a 4",
                    "Cuatro":"3 a 4",
                    "3 a 4":"3 a 4",
                    "Cinco":"5 a 6",
                    "Seis":"5 a 6",
                    "5 a 6":"5 a 6",
                    "Siete":"7 a 8",
                    "Ocho":"7 a 8",
                    "7 a 8":"7 a 8",
                    "Nueve":"9 o más",
                    "Diez":"9 o más",
                    "Once":"9 o más",
                    "Doce o más":"9 o más",
                    "9 o más":"9 o más",
                    np.nan: "unknown"

}

df_intermidate["fami_personashogar"]=df_intermidate["fami_personashogar"].apply(lambda x: personashogar_dict[x])
```

# Imputation Zone

```python
df_intermidate.loc[df_intermidate["punt_ingles"].isna(),"punt_ingles"]=df_intermidate["punt_ingles"].mean()
df_intermidate.loc[df_intermidate["cole_bilingue"].isna(),"cole_bilingue"]="unknown"
df_intermidate.loc[df_intermidate["fami_estratovivienda"].isna(),"fami_estratovivienda"]="Sin Estrato"
df_intermidate.loc[df_intermidate["fami_tieneinternet"].isna(),"fami_tieneinternet"]="unknown"
df_intermidate.loc[df_intermidate["fami_educacionpadre"].isna(),"fami_educacionpadre"]="unknown"
df_intermidate.loc[df_intermidate["fami_educacionmadre"].isna(),"fami_educacionmadre"]="unknown"
df_intermidate.loc[df_intermidate["fami_tieneautomovil"].isna(),"fami_tieneautomovil"]="unknown"
df_intermidate.loc[df_intermidate["fami_tienecomputador"].isna(),"fami_tienecomputador"]="unknown"
df_intermidate.loc[df_intermidate["fami_tienelavadora"].isna(),"fami_tienelavadora"]="unknown"
df_intermidate.loc[df_intermidate["fami_cuartoshogar"].isna(),"fami_cuartoshogar"]="unknown"
df_intermidate.loc[df_intermidate["cole_caracter"].isna(),"cole_caracter"]="NO APLICA"
df_intermidate.loc[df_intermidate["estu_genero"].isna(),"estu_genero"]="unknown"
df_intermidate.loc[df_intermidate["desemp_ingles"].isna(),"desemp_ingles"]=df_intermidate.loc[df_intermidate["desemp_ingles"].isna(),"punt_ingles"].apply(categorize)
```
